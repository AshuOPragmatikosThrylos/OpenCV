```python
import numpy as np                                     
import cv2
import matplotlib.pyplot as plt

grey_img = np.zeros(img.shape[:2])

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        k1 = img[i][j][0]
        k2 = img[i][j][1]
        k3 = img[i][j][2]
        k  = 0.0593*k1 + 0.6780*k2 + 0.2627*k3  #3 formulas from wikipedia....this was just an attempt to realise
                                                #how hard is conversion to greyscale...which is easier by matplotlib cmap
        grey_img[i][j] = k                      #each pixel is stored in BGR format...not  RGB

plt.imshow(grey_img)     
        
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

grey_img = np.zeros((img.shape[0],img.shape[1]))

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        k1 = img[i][j][0]
        k2 = img[i][j][1]
        k3 = img[i][j][2]
        k  = 0.0722*k1 + 0.7152*k2 + 0.2126*k3
        grey_img[i][j] = k
        
plt.imshow(grey_img,cmap="gray")     
        
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

grey_img = np.zeros((img.shape[0],img.shape[1]))

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        k1 = img[i][j][0]
        k2 = img[i][j][1]
        k3 = img[i][j][2]
        k  = 0.114*k1 + 0.587*k2 + 0.299*k3
        grey_img[i][j] = k
        
plt.imshow(grey_img)   #grey store to hogya pr display isliye nhi ho rha kyunki matplotlib alag mapping use krta h.  
                        #how a picture is viewed depends on....captured matrix ,and converted matrix at display
```

```python
for i in range(50):
    for j in range(img.shape[1]):
        img[i][j] = 0
        
for m in range(img.shape[0]):
    for n in range(50):
        img[m][n] = 0
        
for k in range(img.shape[0]-1,img.shape[0]-50,-1):
    for l in range(img.shape[1]):
        img[k][l] = 0
        
for u in range(img.shape[0]):
    for v in range(img.shape[1]-1,img.shape[1]-50,-1):
        img[u][v] = 0
        
cv2.imwrite("Changed_NAO.jpg",img)        
   
```

```python
import cv2
import matplotlib.pyplot as plt

num_rows,num_cols = img.shape[:2]
rot_matrix = cv2.getRotationMatrix2D((num_cols/2,num_rows/2),30,1.0)
img_rotated = cv2.warpAffine(img,rot_matrix,(num_cols,num_rows))

plt.imshow(img_rotated)
```
